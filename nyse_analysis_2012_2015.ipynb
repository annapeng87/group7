```python
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

print(f"Alpaca Key Type: {type(alpaca_api_key)}")
print(f"Alpaca Secret Key Type: {type(alpaca_secret_key)}")
```

```python
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version = "v2"
)
```

```python
timeframe = "1D"

# Set start and end datetimes between now and 5 years ago.
start_date = pd.Timestamp("2012-03-01", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2015-03-01", tz="America/New_York").isoformat()

# Set the ticker information
low_PE_tickers = ['APO','APTS','APTV','BAH','CMRE','COR','DQ','GM','GNRC','HCA','ITT','JKS','MCC','SFUN','WD']

# Get 5 year's worth of historical price data
df_low_PE = alpaca.get_barset(
    low_PE_tickers,
    timeframe,
    start=start_date,
    end=end_date
).df

# Display sample data
df_low_PE.tail()
```

```python
df_closing_prices = pd.DataFrame()
for ticker in low_PE_tickers:
    df_closing_prices[ticker] = df_low_PE[ticker]['close']
    
df_closing_prices.head()
```

```python
df_daily_returns_low_PE = df_closing_prices.pct_change().dropna()
df_daily_returns_low_PE.head()
```

```python
df_daily_returns_low_PE = df_closing_prices.pct_change().dropna()
df_daily_returns_low_PE.head()
```

```python
weights = [1/15]*15
weighted_returns_low_PE = df_daily_returns_low_PE.dot(weights).to_frame()
weighted_returns_low_PE.head()
low_PE_portfolio = weighted_returns_low_PE.rename(columns = {0: 'Low PE Portfolio'})
low_PE_portfolio.index = low_PE_portfolio.index.date
low_PE_portfolio.head()
```

```python
# Set the ticker information
high_PE_tickers = ['AAT','AGRO','AL','ARCO','BKU','ENV','EXPR','FLT','GDOT','GNC','KMI','LPI','ST','TRGP','WSR']

# Get 5 year's worth of historical price data
df_high_PE = alpaca.get_barset(
    high_PE_tickers,
    timeframe,
    start=start_date,
    end=end_date
).df

# Display sample data
df_high_PE.tail()
```

```python
df_closing_prices_high_PE = pd.DataFrame()
for ticker in high_PE_tickers:
    df_closing_prices_high_PE[ticker] = df_high_PE[ticker]['close']
    
df_closing_prices_high_PE.head()
```

```python
df_daily_returns_high_PE = df_closing_prices_high_PE.pct_change().dropna()
df_daily_returns_high_PE.head()
```

```python
weighted_returns_high_PE = df_daily_returns_high_PE.dot(weights).to_frame()
weighted_returns_high_PE.head()
high_PE_portfolio = weighted_returns_high_PE.rename(columns = {0: 'High PE Portfolio'})
high_PE_portfolio.index = high_PE_portfolio.index.date
high_PE_portfolio.head()
```

```python
nyse_csv = Path("Data/nyse_composite.csv")
nyse_df = pd.read_csv(nyse_csv)
```

```python
nyse_df.set_index(pd.to_datetime(nyse_df['Date'], infer_datetime_format = True), inplace = True)
nyse_df.columns
nyse_columns = ["date", "NYSE"]
nyse_df.columns = nyse_columns
nyse_df.drop(['date'], axis = 1, inplace = True)
nyse_df.index = nyse_df.index.date
nyse_df.head()
```

```python
nyse_daily_returns = nyse_df.pct_change().dropna()
nyse_daily_returns.head()
```

```python
three_portfolios_dr = pd.concat([low_PE_portfolio, high_PE_portfolio, nyse_daily_returns], axis = "columns", join = "inner")
three_portfolios_dr.head()
```

```python
three_portfolios_dr.plot(figsize=(20,10), title = "Daily Returns")
```

```python
three_portfolios_cr = (1+three_portfolios_dr).cumprod()
three_portfolios_cr.head()
```

```python
three_portfolios_cr.plot(figsize=(20,10), title = "Cumulative Returns")
```

```python
three_portfolios_std = three_portfolios_dr.std()
three_portfolios_std
```

```python
annualized_std = three_portfolios_dr.std() * np.sqrt(252)
annualized_std
```

```python
three_portfolios_dr.rolling(21).std().plot(figsize=(20,10), title="21 Day Rolling Statistics")
```

```python
sharpe_ratio = (three_portfolios_dr.mean() * 252) / annualized_std
sharpe_ratio
```

```python
sharpe_ratio.plot.bar(title="Sharpe Ratio")
```

```python
correlation = three_portfolios_dr.corr()
correlation
```
